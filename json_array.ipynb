```python
import json
from pprint import pprint
import pandas as pd
```

```python
f = open('india.json', encoding='utf-8')
json_array = []
for line in f:
    try:
        tweet_d = json.loads(line)
        if "info" not in tweet_d:        # avoid the "info" lines from the download files
            json_array.append(tweet_d)
    except:
        pass
```

```python
pprint(json_array[10])
```

```python
print("Total tweets:",len(json_array))
```

```python
list_data = []
for i in json_array:
    if 'retweeted_status' in i:
        list_data.append([i['created_at'], i['user']['screen_name'],i['retweeted_status']['favorite_count'],\
                          i['retweeted_status']['retweet_count'],i['text'],i['user']['location']])
    else:
        list_data.append([i['created_at'], i['user']['screen_name'],0,\
                          i['retweet_count'],i['text'],i['user']['name']])
```

```python
tweet_df = pd.DataFrame(list_data,columns=['created_at', 'screen_name','favorite_count',\
                                           'retweet_count','text','location'])
tweet_df
```

```python
tweet_df.to_csv("India.csv")
```

```python
mybins = [-1, 100, 500, 1000, 2000, 5000, 8000, 50000, 1000000]
fave_groups = tweet_df['favorite_count'].value_counts(bins=mybins)
fave_groups
```

```python
rtwt_groups = tweet_df['retweet_count'].value_counts(bins=mybins)
rtwt_groups
```

```python
big_fave_data = []
for i in json_array:
    if 'retweeted_status' in i and i['retweeted_status']['favorite_count'] > 50000:
        big_fave_data.append(i)
print(len(big_fave_data))
```

```python
rtwt_data = []
for i in json_array:
    if 'retweeted_status' in i:
        rtwt_data.append([i['user']['screen_name'], i['id'],i['retweeted_status']['user']['screen_name'],\
                          i['user']['id'], i['text'],i['created_at']])
print(len(rtwt_data))
```

```python
rtwt_df = pd.DataFrame(rtwt_data,columns=['retweet_screen_name','retweet_id','originator_screen_name',\
                                          'retweet_user_id','text','created_at'])
rtwt_df.to_csv("retweets_India.csv")
```
